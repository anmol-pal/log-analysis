## Populating location record in data frame

```python
import pandas as pd
import re, requests
import time
import json
```

```python
data_file='wrangled_log/only_host_large/part-00000-e3a8b0af-89cb-4326-bf86-018fe738bd2a-c000.csv'
df = pd.read_csv(data_file,delimiter='\t'
                 ,header=0
                 ,dtype='str'
                )
```

```python
df.head(5)
hosts = df['Host'].dropna().unique().tolist()
```

```python
def ip_to_location_batch(ips):
    endpoint = 'http://ip-api.com/batch'
    ips_add = [str(i) for i in ips]
    resp = requests.post(endpoint,data=json.dumps(ips_add), headers={'Content-Type':'application/json'})
    if resp.status_code == 200:
        info = resp.json()
        print(time.ctime(),"OK")
        return info
    else:
        print(time.ctime(),resp.status_code)
        return []
```

```python
hosts = df['Host'].dropna().unique().tolist()
out_hosts = []
country = []
city = []
countryCode = []
lat = []
long = []
regName = []
time_zone = []
if False:
    for i in range(0,len(hosts),100):
        ips = hosts[i:min(i+100,len(hosts))]
        out = ip_to_location_batch(ips)
        if out:
            for o in out:
                if o['status']=='success':
                    out_hosts.append(o['query'])
                    country.append(o['country'])
                    countryCode.append(o['countryCode'])
                    city.append(o['city'])
                    time_zone.append(o['timezone'])
                    lat.append(o['lat'])
                    long.append(o['lon'])
                    regName.append(o['regionName'])
        time.sleep(5)
```

### Storing to a csv countryLargeFineGran.csv

```python
loc_df = pd.DataFrame()
loc_df['Host'] = out_hosts
loc_df['Country'] = country
loc_df['CountryCode'] = countryCode
loc_df['City'] = city
loc_df['Timezone'] = time_zone
loc_df['Lat'] = lat
loc_df['Lon'] = long
loc_df['Reg'] = regName
loc_df.to_csv('countryLargeFineGran.csv')
```
